```python
import pandas as pd
import glob
import os
```

```python
INPUT_FOLDER = "./data"
OUTPUT_FILE = "sampled_flipkart_urls.csv"
SAMPLE_SIZE = 20_000
```

```python
csv_files = glob.glob(os.path.join(INPUT_FOLDER, "*.csv"))
```

```python
all_urls = []

for file in csv_files:
    df = pd.read_csv(file, header=None, names=["url"])
    
    # drop rows with '/hi/' in the URL
    df = df[~df["url"].str.contains("/hi/", case=False, na=False)]

    # randomly sample up to 1 lakh URLs
    sample_df = df.sample(n=min(SAMPLE_SIZE, len(df)), random_state=42)

    all_urls.append(sample_df)
```

```python
merged_df = pd.concat(all_urls, ignore_index=True)
merged_df.to_csv(OUTPUT_FILE, index=False)
```
