```python
import pandas as pd
import math
```

```python
INPUT_FILE = "./sampled_flipkart_urls.csv"
```

```python
df = pd.read_csv(INPUT_FILE)
```

```python
PARTS = 2
```

```python
total = len(df)
size = math.ceil(total / PARTS)
```

```python
for i in range(PARTS):
    start = i * size
    end = start + size
    chunk = df.iloc[start:end]

    outfile = f"./data/sample/urls_part_{i+1}.csv"
    chunk.to_csv(outfile, index=False)
    print(f"Saved {outfile} ({len(chunk)} rows)")
```
