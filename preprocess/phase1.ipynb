````python
import pandas as pd
````

````python
INPUT_FILE_1 = '../data-scraping/product-pages/data/sample-scraped/flipkart_products_scraped_part_1.csv'
INPUT_FILE_2 = '../data-scraping/product-pages/data/sample-scraped/flipkart_products_scraped_part_2.csv'

OUTPUT_FILE = '../dataset/flipkart_products_merged.csv'
````

````python
df_1 = pd.read_csv(INPUT_FILE_1)
df_2 = pd.read_csv(INPUT_FILE_2)
````

````python
df_merged = pd.read_csv(OUTPUT_FILE)
````

````python
df_merged.head()
````

````python
df_merged.info()
````

````python
df_merged = df_merged.drop(columns=['rating_counts', 'review_counts', 'discounted_price'])
````

````python
df_merged['price'] = (
    df_merged['price']
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)
````

````python
df_merged['discount'] = (
    df_merged['discount']
    .str.replace('% off', '', regex=False)
    .astype(float)
)
````

````python
df_merged.isnull().sum()
````

````python
df_merged = df_merged.dropna(subset=['price', 'discount'])
````

````python
df_merged = df_merged.dropna(subset=['title'])
````

````python
df_merged['description'].fillna('No Description', inplace=True)
````

````python
df_merged['hierarchy'].fillna('Unknown', inplace=True)
````

````python
df_merged.isnull().sum()
````

````python
df_merged.info()
````

````python
df_merged["hierarchy_list"] = df_merged["hierarchy"].apply(
    lambda x: [p.strip() for p in x.split(">")] if pd.notna(x) else None
)
````

````python
df_merged["hierarchy_depth"] = df_merged["hierarchy_list"].apply(lambda x: len(x) if x else 0)
````

````python
df_merged["hierarchy_depth"].describe()
````

````python
max_depth = 8
````

````python
for i in range(max_depth):
    df_merged[f"level_{i+1}"] = df_merged["hierarchy_list"].apply(
        lambda x: x[i] if x and len(x) > i else None
    )
````

````python
import re
````

````python
emoji_pattern = re.compile(
    "[" 
    "\U0001F600-\U0001F64F"  # emoticons
    "\U0001F300-\U0001F5FF"  # symbols & pictographs
    "\U0001F680-\U0001F6FF"  # transport
    "\U0001F1E0-\U0001F1FF"  # flags
    "\U00002700-\U000027BF"  # dingbats
    "\U0001F900-\U0001F9FF"  # supplemental symbols
    "\U0001FA70-\U0001FAFF"  # extended symbols
    "\U00002600-\U000026FF"  # misc symbols
    "\U00002B00-\U00002BFF"  # arrows
    "]+",  
    flags=re.UNICODE
)
````

````python
def remove_emojis(text):
    if pd.isna(text):
        return text
    return emoji_pattern.sub('', text)
````

````python
df_merged["title"] = df_merged["title"].apply(remove_emojis)
df_merged["description"] = df_merged["description"].apply(remove_emojis)
````
